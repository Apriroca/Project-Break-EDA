```python
import matplotlib.pyplot as plt
import seaborn as sns 
# Configuración estética
sns.set_theme(style="whitegrid")
```

```python
# 1. Analisis por Barrio (Venta vs Alquiler)
#Agrupamos por barrio y operacion para ver el precio medio por m2 
barrios_stats = df_limpio.groupby(['Barrio', 'Operación'])['Precio unitario'].median().unstack().reset_index()

# 15 barrios mas caros en venta
plt.figure(figsize=(12,6))
top_barrios_venta = barrios_stats.sort_values(by='Venta', ascending=False).head(15)
sns.barplot(data=top_barrios_venta, x='Venta', y='Barrio', palette='viridis')
plt.title('Top 15 Barrios con mayor precio medio por m2 (Venta)')
plt.xlabel('Precio medio (€/m2)')
plt.show()
```

```python
# Visualizacion de los 15 barrios mas caros en Alquiler
plt.figure(figsize=(12, 6))
top_barrios_alquiler = barrios_stats.sort_values(by='Alquiler', ascending=False).head(15)
sns.barplot(data=top_barrios_alquiler, x='Alquiler', y='Barrio', palette='magma')
plt.title('Top 15 Barrios con mayor precio medio por m2 (Alquiler)')
plt.xlabel('Precio medio (€/m2)')
plt.show()
```

```python
#  Graficamos el precio promedio por cada rango de m2 para venta
# Creamos rangos de superficie para simplificar
# Por ejemplo: Pequeño, Mediano, Grande, Muy Grande
bins = [0, 50, 100, 150, 200, 300, 1000]
labels = ['0-50', '50-100', '100-150', '150-200', '200-300', '300+']
df_limpio['Rango_Superficie'] = pd.cut(df_limpio['Superficie'], bins=bins, labels=labels)

plt.figure(figsize=(10, 6))
sns.barplot(data=df_limpio[df_limpio['Operación'] == 'Venta'], 
            x='Rango_Superficie', y='Precio', palette='Blues_d')

plt.title('Precio Promedio según el Tamaño de la Vivienda (Venta)')
plt.xlabel('Superficie (m²)')
plt.ylabel('Precio Medio (€)')
plt.show()
```

```python
# 4. Análisis de Tension (Rentabilidad bruta por Distrito)
# Calculamos el precio medio de venta y alquiler por distrito
yield_analysis = df_limpio.groupby(['Distrito', 'Operación'])['Precio'].mean().unstack()

# Calculamos la rentabilidad Bruta Anual (%)
# Nota: Asumimos que los distritos existen en ambos DataFrames
yield_analysis['Yield_%'] = (yield_analysis['Alquiler'] * 12 / yield_analysis['Venta']) * 100

# Visualización
plt.figure(figsize=(12, 8))
yield_sorted = yield_analysis.sort_values(by='Yield_%', ascending=False)
sns.barplot(x=yield_sorted['Yield_%'], y=yield_sorted.index, palette='RdYlGn_r')
plt.axvline(x=5, color='red', linestyle='--', label='Media mercado (Ref)')
plt.title('Tensión de Mercado: Rentabilidad Bruta por Distrito (%)')
plt.xlabel('Rentabilidad Bruta Anual (%) - A mayor %, más tensionado/caro el alquiler respecto a compra')
plt.legend()
plt.show()

#Interpretacion de los resultados esperados:
#Brechas Geograficas: Veras que distritos como L'Eixample o Ciutat Vella suelen duplicar o triplicar el precio por m2 de barrios periféricos.
#Correlacion: En ventas, la Superficie suele tener una correlacion muy alta con el Precio. En alquiler, a veces el numero de Dormitorios pesa más que los metros cuadrados debido a la demanda de pisos compartidos.
#Tension de Alquiler: Si encuentras distritos con un Yield superior al 7-8%, son zonas donde el alquiler está proporcionalmente mucho más caro que la compra, indicando una alta presion de demanda de alquiler (zonas tensionadas) o precios de compra bajos por riesgo percibido.
```

```python
# Función para clasificar variables por tipo
def clasificar_variables(df, umbral_categoria=10, umbral_continua=30):
    """
    Clasifica las variables del DataFrame según su cardinalidad.
    """
    df_temp = pd.DataFrame({
        'Card': df.nunique(),
        '%_Card': (df.nunique() / len(df) * 100).round(2),
        'Tipo': df.dtypes
    })
    
    # Corrección para cuando solo hay un valor
    df_temp.loc[df_temp['Card'] == 1, '%_Card'] = 0.00
    
    # Clasificar
    df_temp['tipo_sugerido'] = 'Categorica'
    df_temp.loc[df_temp['Card'] == 2, 'tipo_sugerido'] = 'Binaria'
    df_temp.loc[df_temp['Card'] >= umbral_categoria, 'tipo_sugerido'] = 'Numerica discreta'
    df_temp.loc[df_temp['%_Card'] >= umbral_continua, 'tipo_sugerido'] = 'Numerica continua'
    
    return df_temp

# Clasificar variables
df_tipos = clasificar_variables(df)
df_tipos
```

```python
# Función para graficar distribuciones categóricas
def pinta_distribucion_categoricas(df, columnas_categoricas, relativa=False, mostrar_valores=False):
    num_columnas = len(columnas_categoricas)
    num_filas = (num_columnas // 2) + (num_columnas % 2)
    
    fig, axes = plt.subplots(num_filas, 2, figsize=(15, 5 * num_filas))
    axes = axes.flatten()
    
    for i, col in enumerate(columnas_categoricas):
        ax = axes[i]
        serie = df[col].value_counts()
        
        if relativa:
            serie = serie / serie.sum()
            ylabel = 'Frecuencia Relativa'
        else:
            ylabel = 'Frecuencia'
        
        sns.barplot(x=serie.index, y=serie.values, ax=ax, palette='viridis', hue=serie.index, legend=False)
        ax.set_ylabel(ylabel)
        ax.set_title(f'Distribución de {col}')
        ax.set_xlabel('')
        ax.tick_params(axis='x', rotation=45)
        
        if mostrar_valores:
            for p in ax.patches:
                height = p.get_height()
                formato = f'{height:.2%}' if relativa else f'{height:,.0f}'
                ax.annotate(formato, (p.get_x() + p.get_width() / 2., height),
                           ha='center', va='bottom', fontsize=9)
    
    for j in range(i + 1, num_filas * 2):
        axes[j].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
# Frecuencias absolutas y relativas
for cat in categoricas:
    print(f"{'='*50}")
    print(f"{cat.upper()}")
    print(f"{'='*50}")
    freq_abs = df_limpio[cat].value_counts()
    freq_rel = (freq_abs / len(df_limpio) * 100).round(2)
    resumen = pd.DataFrame({'Absoluta': freq_abs, 'Relativa (%)': freq_rel})
    print(resumen.to_string())
    print("\n")
```

```python
# Gráficos - Variables categóricas principales
pinta_distribucion_categoricas(df_limpio, ['Operación', 'Fuente', 'Anunciante', 'Demanda'], mostrar_valores=True)
```

```python
# Gráficos con frecuencias relativas (%)
pinta_distribucion_categoricas(df_limpio, ['Operación', 'Fuente', 'Anunciante', 'Demanda'], 
                                relativa=True, mostrar_valores=True)
```

```python
# Porcentaje de Sí para cada amenity
print("AMENITIES - % de inmuebles con cada característica")
print("="*50)

amenities_pct = {}
for amenity in binarias:
    pct_si = (df_limpio[amenity] == 'Sí').sum() / len(df_limpio) * 100
    amenities_pct[amenity] = pct_si
    print(f"{amenity:<20}: {pct_si:>5.1f}%")

# Ordenar de mayor a menor
amenities_sorted = dict(sorted(amenities_pct.items(), key=lambda x: x[1], reverse=True))
```

```python
# Gráfico de amenities
fig, ax = plt.subplots(figsize=(10, 6))

amenities_df = pd.Series(amenities_sorted)
bars = ax.barh(amenities_df.index, amenities_df.values, color='teal')
ax.set_xlabel('% de inmuebles')
ax.set_title('Disponibilidad de Amenities')
ax.invert_yaxis()

for bar in bars:
    ax.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
            f'{bar.get_width():.1f}%', va='center')

plt.tight_layout()
plt.show()
```
